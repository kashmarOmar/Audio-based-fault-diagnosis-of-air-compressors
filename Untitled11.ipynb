```python
import os
import requests
import zipfile
import tempfile

dataset_url = 'https://www.mathworks.com/supportfiles/audio/AirCompressorDataset/AirCompressorDataset.zip'
download_folder = os.path.join(tempfile.gettempdir(), 'AirCompressorDataSet')

if not os.path.exists(download_folder):
    os.makedirs(download_folder)

response = requests.get(dataset_url)

if response.status_code == 200:
    zip_file_path = os.path.join(download_folder, 'AirCompressorDataset.zip')

    with open(zip_file_path, 'wb') as f:
        f.write(response.content)

    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(download_folder)

    os.remove(zip_file_path)
    import librosa
import librosa.display

# Define the path to the dataset location
dataset_location = os.path.join(tempfile.gettempdir(), 'AirCompressorDataSet', 'AirCompressorDataset')

# Create an empty list to store audio files and their labels
audio_data = []
labels = []

# Loop through subfolders and collect audio data
for root, dirs, files in os.walk(dataset_location):
    for file in files:
        if file.endswith(".wav"):  # Adjust the file extension as needed
            file_path = os.path.join(root, file)
            audio_data.append(file_path)
            labels.append(os.path.basename(root))

# Now you can work with the audio_data list and labels to process your audio dataset.
```

```python
from collections import Counter
import pandas as pd
import numpy as np

# Assuming you have the 'labels' list from the previous code
label_counts = Counter(labels)

# Print the count of each category
for label, count in label_counts.items():
    print(f'{label}: {count}')
```

```python
from sklearn.model_selection import train_test_split
import random

# Assuming you have a list of audio file paths 'audio_data' and their corresponding labels 'labels'
# Shuffle the dataset
audio_data_shuffled, labels_shuffled = zip(*random.sample(list(zip(audio_data, labels)), len(audio_data)))

# Split the shuffled dataset into training and testing sets (70% train, 30% test)
x_train, x_test, y_train, y_test = train_test_split(
    audio_data_shuffled, labels_shuffled, test_size=0.3, random_state=0)

# 'x_train' and 'y_train' now contain the training data and labels,
# and 'x_test' and 'y_test' contain the test data and labels.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have 'y_train' and 'y_test' containing labels for the training and test sets
# Calculate unique labels and their counts
distinct_labels = pd.unique(y_train)
train_label_counts = pd.Series(y_train).value_counts().reindex(distinct_labels).fillna(0)
test_label_counts = pd.Series(y_test).value_counts().reindex(distinct_labels).fillna(0)

# Create a DataFrame for plotting
label_distribution_df = pd.DataFrame({'Labels': distinct_labels, 'Train Counts': train_label_counts, 'Test Counts': test_label_counts})

# Create a stacked bar plot
ax = label_distribution_df[['Train Counts', 'Test Counts']].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Labels')
plt.ylabel('Count')
plt.title('Label Distribution in Training and Test Sets')
plt.legend(["Train Counts", "Test Counts"], loc="best")

plt.show()
```

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set your desired sample rate
sample_rate = 16000

# Generate random indices for audio samples
num_samples_to_display = min(8, len(x_train))  # Ensure not more samples than available
random_indices = random.sample(range(len(x_train)), num_samples_to_display)

# Create subplots for 3D spectrogram plots
plt.figure(figsize=(10, 10))
for n, random_index in enumerate(random_indices, start=1):
    audio_file = x_train[random_index]

    # Load the audio file
    audio_data, _ = sf.read(audio_file)  # Read the audio file

    # Compute the short-time Fourier transform (STFT)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(audio_data)), ref=np.max)

    # Create a 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    time_axis, frequency_axis = np.meshgrid(np.arange(D.shape[1]), np.arange(D.shape[0]))
    ax.plot_surface(time_axis, frequency_axis, D, cmap='viridis')

    # Set plot labels
    ax.set_xlabel('Time')
    ax.set_ylabel('Frequency')
    ax.set_zlabel('Amplitude (dB)')

    # Use the corresponding label for the title
    ax.set_title(f'3D Spectrogram - {y_train[random_index]}')  # Use y_train instead of unique_labels

plt.tight_layout()
plt.show()
```
